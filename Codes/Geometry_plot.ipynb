```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d.art3d import Poly3DCollection

# Create the figure and 3D axis
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Square parameters (in pixels and meters)
square_size = 20  # 20x20 square
half_size = square_size / 2

# Define the square in the xy plane (z=0)
square_corners = np.array([[-half_size, -half_size, 0],
                           [half_size, -half_size, 0],
                           [half_size, half_size, 0],
                           [-half_size, half_size, 0],
                           [-half_size, -half_size, 0]])  # To close the loop

# Plot the edges of the square on top
ax.plot(square_corners[:, 0], square_corners[:, 1], square_corners[:, 2], 'k-', lw=2)

# Point at the center of the square
center_point = np.array([0, 0, 10])  # Height of 10 units in the z-axis
center_point2 = np.array([0, 0, 0])  # Height of 10 units in the z-axis
# Plot the center point in the 3D space
ax.scatter(center_point[0], center_point[1], center_point[2], color='r', s=100)

# Line normal to the plane (z-axis line from the point to the plane)
normal_line = np.array([[0, 0, 0], center_point])
ax.plot(normal_line[:, 0], normal_line[:, 1], normal_line[:, 2], 'b--', lw=2)

# Line from the point to the middle of an edge (midpoint of the right edge)
mid_edge_point = np.array([half_size, 0, 0])  # Midpoint of the right edge
connecting_line = np.array([mid_edge_point, center_point])
ax.plot(connecting_line[:, 0], connecting_line[:, 1], connecting_line[:, 2], 'g-', lw=2)

# Add a solid black line from the center to the edge at (10, 0)
edge_point = np.array([10, 0, 0])  # Point at (10, 0, 0)
ax.plot([center_point2[0], edge_point[0]], 
        [center_point2[1], edge_point[1]], 
        [center_point2[2], edge_point[2]], 
        color='black', lw=2)

# Label the line 'X'
ax.text(edge_point[0] -8, edge_point[1]+2, edge_point[2] , 'X', fontsize=12, color='black')

# Add annotation for theta
theta_text_position = (center_point + mid_edge_point) / 2 - 3  # Midpoint of the two lines
ax.text(theta_text_position[0] + 0.5, theta_text_position[1], theta_text_position[2] + 5.5, r'$\theta$', fontsize=15)

# Add annotation for h
ax.text(theta_text_position[0] - 2.5, theta_text_position[1], theta_text_position[2] + 1.5, r'h', fontsize=15)

# Label axes
ax.set_xlabel('X (pixels)')
ax.set_ylabel('Y (pixels)')
ax.set_zlabel('Z (height)')

# Set limits for a better view
ax.set_xlim([-10, 10])
ax.set_ylim([-10, 10])
ax.set_zlim([0, 10])

# Add legend
ax.legend()

# Display the plot
plt.show()
```
